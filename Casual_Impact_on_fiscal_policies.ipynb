```python
from causalimpact import CausalImpact
import pandas as pd
```

```python
data_1 = pd.read_csv('/content/drive/MyDrive/filled_data.csv')
```

```python
data_1
```

```python
data = data_1.iloc[: , [3, 0]].copy()
```

```python
data
```

```python
data.plot()
```

```python
#define the year in which the policy or impact started ==> here 12 represents 2007
pre_period = [0,12]
post_period = [15,26]
```

```python
impact = CausalImpact(data, pre_period, post_period)
```

```python
impact.run()
```

```python
impact.plot()
```

```python
impact.inferences
```

```python
impact.summary(output="report")
```

```python
impact.results.results.plot_diagnostics()
```

```python
impact.results.results.plot_components()
```
