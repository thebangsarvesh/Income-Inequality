```python
import pandas as pd

data = pd.read_excel("/content/filled_combined_data.xlsx")
data.head()
```

```python
data_filled = data.fillna(data.mean())

print(data_filled.isnull().sum())
```

```python
data_filled.head()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# the columns to analyze, i asked chatgpt which columns were likely to be impacted most.
# gst came in 2017 so thts the year column

columns_to_analyze = [
    "expenditure, percent of GDP (% of GDP)",
    "Corporate Tax ",
    "Personal Income Tax@ ",
    "Income tax collected",
    "% Share of  Agriculture and Allied sector to total GDP(curr)"
]

results={}

for column in columns_to_analyze:
    results[column] = {}

    #splitting the data into pre-GST and post-GST
    pre_gst = data_filled[data_filled["Year"] < 2017][column]
    post_gst = data_filled[data_filled["Year"] >= 2017][column]

    #fit a linear regression model to the pre-GST data
    pre_gst_model = LinearRegression()
    pre_gst_model.fit(np.array(pre_gst.index).reshape(-1, 1), pre_gst)

    #fit a linear regression model to the post-GST data
    post_gst_model.fit(np.array(post_gst.index).reshape(-1, 1), post_gst)

    # here after applying linear regreession , i got the trend if change before and after gst.
    results[column]["pre_gst_slope"] = pre_gst_model.coef_[0]
    results[column]["post_gst_slope"] = post_gst_model.coef_[0]

    #foraecasting the post-GST trend for 2023-2026 using SARIMA
    sarima_model = SARIMAX(post_gst, order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
    sarima_model_fit = sarima_model.fit(disp=False)
    forecast = sarima_model_fit.predict(len(post_gst), len(post_gst) + 3)
    post_gst = post_gst.append(forecast)

    ##Plots
    plt.figure(figsize=(10, 6))
    plt.plot(pre_gst, label="Pre-GST")
    plt.plot(post_gst, label="Post-GST with forecast")
    plt.xlabel("Year")
    plt.ylabel(column)
    plt.title(f"Trend Analysis and forecast for {column}")
    plt.legend()
    plt.grid(True)
    plt.show()

 #differnces between trend
for column, result in results.items():
    print(f"{column}:")
    print(f"  Pre-GST trend slope: {result['pre_gst_slope']}")
    print(f"  Post-GST trend slope: {result['post_gst_slope']}")
    print(f"  Difference in slope: {result['post_gst_slope'] - result['pre_gst_slope']}")
    if result['post_gst_slope'] > result['pre_gst_slope']:
        print("  GST had a positive impact.\n")
    elif result['post_gst_slope'] < result['pre_gst_slope']:
        print(" GST had a negative impact.\n")
    else:
        print("GST had a neutral impact.\n")
```

```python
import pandas as pd
from causalimpact import CausalImpact

# Load your data
data = pd.read_excel('filled_combined_data.xlsx')
data.set_index('Year', inplace=True)

# Move the response variable to the first column
cols = data.columns.tolist()
cols.remove('expenditure, percent of GDP (% of GDP)')
cols = ['expenditure, percent of GDP (% of GDP)'] + cols
data = data[cols]

# Reindex the DataFrame to include all years
all_years = pd.date_range(start=min(data.index), end=max(data.index), freq='YS')
data = data.reindex(all_years)

# Define your intervention point
pre_period = ['1995', '2016']
post_period = ['2017', '2022']

# Run the analysis
impact = CausalImpact(data, pre_period, post_period)

# Print and plot the results
print(impact.summary())
impact.plot()
```
